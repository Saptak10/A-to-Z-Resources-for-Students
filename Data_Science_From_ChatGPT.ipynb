Pandas

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)

# Filtering Data
filtered_df = df[df['Age'] > 25]
print("Filtered DataFrame:\n", filtered_df)

# Handling Missing Data
df['Salary'] = [50000, None, 60000]
df_filled = df.fillna(0)
print("DataFrame with missing values filled:\n", df_filled)
```

Scikit-Learn

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X, y = data.data, data.target

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train a RandomForest model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, predictions)}")
```

Statistics and Probability

1. Descriptive Statistics:

Mean: The average value of a dataset.

Median: The middle value when the data is sorted.

Mode: The most frequently occurring value in the dataset.

Variance: Measures the spread of the data points around the mean.

Standard Deviation: The square root of the variance, indicating how much the data deviates from the mean.

Range: The difference between the maximum and minimum values.

Quartiles and Interquartile Range (IQR): Divide the data into quarters; IQR measures the spread of the middle 50% of the data.

```python
import numpy as np

data = [10, 12, 15, 14, 18, 20, 25, 30, 18]

# Mean, Median, Mode
mean = np.mean(data)
median = np.median(data)
mode = max(set(data), key=data.count)

# Variance and Standard Deviation
variance = np.var(data)
std_dev = np.std(data)

# Range and IQR
range_data = np.ptp(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)

print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
print(f"Variance: {variance}, Standard Deviation: {std_dev}")
print(f"Range: {range_data}, IQR: {iqr}")
```

2. Data Distributions:

Normal Distribution: Also known as the Gaussian distribution, it's a bell-shaped curve where most of the data points are centered around the mean.

Skewness: Indicates the asymmetry of the distribution. Positive skew means a long tail on the right; negative skew means a long tail on the left.

Kurtosis: Measures the "tailedness" of the distribution. High kurtosis indicates heavy tails, while low kurtosis indicates light tails.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotting the distribution
sns.histplot(data, kde=True)
plt.title("Normal Distribution")
plt.show()
```

3. Correlation and Covariance:

Correlation: Measures the strength and direction of a linear relationship between two variables, typically ranging from -1 to 1.

Covariance: Indicates the direction of the linear relationship between variables. A positive covariance means that as one variable increases, the other tends to increase.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Covariance
covariance = np.cov(x, y)[0, 1]

# Correlation
correlation = np.corrcoef(x, y)[0, 1]

print(f"Covariance: {covariance}, Correlation: {correlation}")
```

B. Probability Theory
1. Basics of Probability:

Probability: A measure of the likelihood that an event will occur, ranging from 0 (impossible) to 1 (certain).

Conditional Probability: The probability of an event occurring given that another event has already occurred.

Bayes' Theorem: A way to update the probability of an event based on new evidence.

```python
# Example of basic probability
probability_of_rain = 0.3
probability_of_not_rain = 1 - probability_of_rain

# Example of conditional probability using Bayes' Theorem
# P(A|B) = (P(B|A) * P(A)) / P(B)

# Let's assume:
# P(A) = Probability of having a disease = 0.01
# P(B|A) = Probability of testing positive given the disease = 0.99
# P(B) = Probability of testing positive = 0.05

P_A = 0.01
P_B_given_A = 0.99
P_B = 0.05

P_A_given_B = (P_B_given_A * P_A) / P_B
print(f"Probability of having the disease given a positive test result: {P_A_given_B:.2f}")
```
